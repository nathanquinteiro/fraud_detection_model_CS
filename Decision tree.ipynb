```python
import pandas as pd
from sklearn.metrics import f1_score, precision_score, recall_score
```

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from sklearn import tree
```

```python
train_data = pd.read_csv("../final_train.csv")
train_data.head()
```

```python
msk = np.random.rand(len(train_data)) < 0.8
train = train_data[msk]
val = train_data[~msk]
```

```python
len(train)
```

```python
len(val)
```

```python
train_features = train.drop(["Unnamed: 0", "suspicious"], axis=1).values
train_labels = train["suspicious"].values
```

```python
val_features = val.drop(["Unnamed: 0", "suspicious"], axis=1).values
val_labels = val["suspicious"].values
```

```python
clf = tree.DecisionTreeRegressor()
clf = clf.fit(train_features, train_labels)
```

```python
predictions = clf.predict(val_features)
predictions = np.array(list(map(lambda x: 0 if x < 0.2 else 1, predictions)))
```

```python
predictions
```

```python
sum(val_labels == predictions) / len(predictions)
```

```python
f1_score(val_labels, predictions)
```

```python
recall_score(val_labels, predictions)
```

```python
precision_score(val_labels, predictions)
```

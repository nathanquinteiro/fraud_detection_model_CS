```python
import pandas as pd
from sklearn.metrics import f1_score, precision_score, recall_score
```

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
```

```python
train_data = pd.read_csv("../final_train.csv")
train_data.head()
```

```python
msk = np.random.rand(len(train_data)) < 0.8
train = train_data[msk]
val = train_data[~msk]
```

```python
len(train)
```

```python
len(val)
```

```python
train_features = train.drop(["Unnamed: 0", "suspicious"], axis=1)
train_labels = train["suspicious"]
```

```python
val_features = val.drop(["Unnamed: 0", "suspicious"], axis=1)
val_labels = val["suspicious"]
```

```python
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 100, max_depth=20, random_state = 42, n_jobs=4)
```

```python
# Train the model on training data
rf.fit(train_features, train_labels);
```

```python
predictions = rf.predict(val_features)
predictions = list(map(lambda x: 0 if x < 0.5 else 1, predictions))
```

```python
predictions[:10]
```

```python
val_labels.head(10)
```

```python
sum(val_labels == predictions) / len(predictions)
```

```python
f1_score(val_labels, predictions)
```

```python
recall_score(val_labels, predictions)
```

```python
precision_score(val_labels, predictions)
```
